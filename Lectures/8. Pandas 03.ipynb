## 3. Extra Work

```python
price_AAPL = yf.download('AAPL', start=start, end=end)
```

```python
price_AAPL.columns = price_AAPL.columns.droplevel(1)
price_AAPL.head()
```

```python
price_AAPL['daily_diffrence'] = price_AAPL['Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_diffrence'] > 0, 1, 0)
price_AAPL.head(10)
```

```python
price_AAPL['position'] = price_AAPL['signal'].diff()
price_AAPL.head(10)
```

```python
price_AAPL[price_AAPL['position']==1].index
```

```python
price_AAPL.loc[price_AAPL['position']==1, 'Close']
```

```python
plt.figure(figsize=(15,8))

plt.plot(price_AAPL['Close'], c = 'r', lw = 2)
plt.plot(price_AAPL[price_AAPL['position']==1].index,
         price_AAPL.loc[price_AAPL['position']==1, 'Close'], 'o', c = 'g', markersize = 7
        )

plt.plot(price_AAPL[price_AAPL['position']==-1].index,
         price_AAPL.loc[price_AAPL['position']==-1, 'Close'], 'x', c = 'k', markersize = 7
        )

plt.grid()
plt.show()
```

```python
import wbdata
```

```python
countries = ['USA', 'CHN']
indicator = {"NY.GDP.PCAP.CD": "GDP per capita", "SP.DYN.LE00.IN": "Life expectancy"}
data_date = (pd.to_datetime('2010'), pd.to_datetime('2020'))

data = wbdata.get_dataframe(indicator, country=countries)
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data['Log GDP per capita'] = np.log(data['GDP per capita'])
```

```python
data.head()
```

```python
X = sm.add_constant(data['Log GDP per capita'])
y = data['Life expectancy']

ols_model = sm.OLS(y, X).fit()
print(ols_model.summary())
```

```python
data.columns = ['GDP_per_capita', 'Life_expectancy', 'Log_GDP_per_apita']
```

```python
data.head()
```

```python
model = smf.ols('Life_expectancy ~ np.log(GDP_per_capita)', data).fit()
print(model.summary())
```
